```python
class Monster:
    def __init__(self, name, attack, defense):
        self.name = name
        self.attack = attack
        self.defense = defense
    def info(self):
        print(f'이름: {self.name}, 공격력: {self.attack}, 방어력: {self.defense}')

shark = Monster("상어", 100, 50)
shark.info()
wolf = Monster("늑대", 200, 30)
wolf.info()
baron = Monster("바론", 300, 20)
baron.info()
```

```python
# 캐릭터 클래스를 만들고, 속성 메서드 사용해보기
class Character:
    def __init__(self, name, level, job, hp):
        self.name = name
        self.level = level
        self.job = job 
        self.hp = hp
    def info(self):
        print('------캐릭터 정보------') 
        print(f'이름: {self.name}')
        print(f'직업: {self.job}')
        print(f'체력: {self.hp}') 
        
char1 = Character("캐릭1", 200, "히어로", 50000) 
char2 = Character("캐릭2", 200, "팔라딘", 70000)
char3 = Character("캐릭3", 200, "나이트로드", 30000)
char4 = Character("캐릭4", 200, "다크나이트", 40000)
char5 = Character("캐릭5", 200, "아크메이지(불,독)", 60000)
char6 = Character("캐릭6", 200, "아크메이지(썬,콜)", 60000)
char7 = Character("캐릭7", 200, "비숍", 80000)
char8 = Character("캐릭8", 200, "에반", 90000)
char9 = Character("캐릭9", 200, "카이저", 100000)
char10 = Character("캐릭10", 200, "캡틴", 110000) 

char1.level += 10 #레벨업
char1.hp += 10000  #체력업
char1.info()    #캐릭터 정보 출력
char1.say = "레이징 블로우"
print(char1.say) 


import pandas as pd 
df = pd.DataFrame([char1.__dict__, char2.__dict__, char3.__dict__, char4.__dict__, char5.__dict__, char6.__dict__, char7.__dict__, char8.__dict__, char9.__dict__, char10.__dict__])
df
```

```python
import requests  
from bs4 import BeautifulSoup as bs

response = requests.get("https://www.naver.com")  # 네이버 접속
print(response.text) # html 코드를 가져옴 #.text가 속성 #response는 객체 #response.text는 메서드
#response.text는 문자열이다.
```

```python
html = response.text
soup = bs(html, 'html.parser') #html.parser는 파서
tag = soup.select_one("선택자") #선택자는 html에서 가져올 태그를 지정
tag = soup.select("선택자") #태그 여러개 선택자는 html에서 가져올 태그를 지정
```
