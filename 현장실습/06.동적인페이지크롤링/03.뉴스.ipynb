```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
import time

driver = webdriver.Chrome("c://chromedriver.exe")
driver.get("https://news.daum.net/economic")
time.sleep(2)
```

```python
articles = driver.find_elements(By.CSS_SELECTOR, ".list_realtime > li")

for article in articles:
    t_time = article.find_element(By.CSS_SELECTOR, ".txt_time").text
    press = article.find_element(By.CSS_SELECTOR, ".info_cp").text
    anchor = article.find_element(By.CSS_SELECTOR, ".link_txt")
    title = anchor.text
    link = anchor.get_attribute("href")
    print(t_time, press, title, link)
```

```python
# 네트워크 분석 - ajax 크롤링
import requests
import json

response = requests.get("https://news.daum.net/api/harmonydic/contents/news.json?category=economic&approved=true&page=1&pageSize=20&pagesToShow=10&range=1")

# 문자열 json을 딕셔너리로 변환
data = json.loads(response.text) 
articles = data['list']
for article in articles:
    link = article['contentUrl']
    press = article['cpKorName']
    title = article['title']
    cotents = article['summary']
    print(press, title, link, cotents)
```

```python
for i in range(1, 4):
    response = requests.get(f"https://news.daum.net/api/harmonydic/contents/news.json?category=economic&approved=true&page={i}&pageSize=20&pagesToShow=10&range=1")
    data = json.loads(response.text)
    articles = data['list']
    for article in articles:
        link = article['contentUrl']
        press = article['cpKorName']
        title = article['title']
        cotents = article['summary']
        print(press, title, link, cotents)
```
